```python
import ee
import geemap

import geopandas as gpd
import matplotlib as plt
import shapely.geometry

from rembg import remove

from PIL import Image
```

```python
ee.Authenticate()
ee.Initialize()
```

```python
file = '/content/drive/MyDrive/datasets/roi/Area_do_Imovel.shp'
geo = gpd.read_file(file, rows=1)
geo.plot()
```

```python
coords = list(shapely.geometry.mapping(geo['geometry'])['features'][0]['geometry']['coordinates'][0])
roi = ee.Geometry.Polygon(coords)
roi
```

```python
collection = ee.ImageCollection("COPERNICUS/S2_SR_HARMONIZED")\
                .filterBounds(roi)\
                .filterDate('2023-01-29', '2023-07-06')\
                .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 10))
collection
```

```python
img_sel = collection.select('B.*').sort('CLOUDY_PIXEL_PERCENTAGE').first()
img_sel
```

```python
Map = geemap.Map()
Map.centerObject(roi, 14)
img = img_sel
ndvi = img.normalizedDifference(['B8', 'B4']).rename('NDVI')
ndwi = img.normalizedDifference(['B3', 'B8']).rename('NDWI')
Map.addLayer(img, {'bands': ['B4', 'B3', 'B2'], 'min': 100, 'max': 1500}, 'S2 RGB')
Map.addLayer(img, {'bands': ['B11', 'B8', 'B4'], 'min': 0, 'max': 4000, 'gamma': 1.5}, 'Sw NIR')
Map.addLayer(img, {'bands': ['B8', 'B4', 'B3'], 'min': 0, 'max': 4000, 'gamma': 1.5}, 'NIR RED')
Map.addLayer(ndvi, {'palette': ['red', 'yellow', 'green'], 'min': -1, 'max': 1}, 'NDVI')
Map.addLayer(ndwi, {'palette': ['red', 'yellow', 'green', 'cyan', 'blue'], 'min': -1, 'max': 1}, 'NDWI')
Map.addLayer(roi, {}, 'Área')
Map
```

```python
# Definindo áre de vegetação densa NDVI>= 0.8
ndwiThreshold = ndvi.gt(0.8)
ndvimask = ndwiThreshold.updateMask(ndwiThreshold)

Map = geemap.Map()
Map.centerObject(roi, 14)

Map.addLayer(ndvimask , {'palette': ['white', 'green'], 'min': 0.8, 'max': 1}, 'NDVI')
Map.addLayer(img, {'bands': ['B4', 'B3', 'B2'], 'min': 100, 'max': 1500}, 'S2 RGB')
Map.addLayer(roi, {}, 'Área')
Map
```

```python
stat = ndvi.reduceRegion(**{
    'reducer': ee.Reducer.median().combine(**{
        'reducer2': ee.Reducer.mean(),
        'sharedInputs': True
    }).combine(**{
        'reducer2': ee.Reducer.max(),
        'sharedInputs': True
    }).combine(**{
        'reducer2': ee.Reducer.min(),
        'sharedInputs': True
    }),
    'geometry': roi,
    'scale': 10,
    'maxPixels': 1e13
})
stat.getInfo()
```

```python
ndvimask
```

```python
#Verificando a área de vremanescente florestal com NDVI>=0.8
vnpoly = ndvimask.select('NDVI').reduceToVectors(**{
  'geometry': roi,
  'crs': roi.projection(),
  'scale': 10,
  'geometryType': 'polygon',

});
vnpoly
```

```python
Map = geemap.Map()
Map.centerObject(roi, 14)


Map.addLayer(img, {'bands': ['B4', 'B3', 'B2'], 'min': 100, 'max': 1500}, 'S2 RGB')
area = ee.FeatureCollection(roi)
# Paint FeatureCollection to an image using collection-wide style arguments.
area = ee.Image().paint(area, color='cyan', width=2)\
    .toByte()

#Map.addLayer(area, {'palette': ['orange'], 'max': 3, 'opacity': 0.5}, 'Area')
Map.addLayer(area, {'palette':'orange'}, 'Area')
Map.addLayer(vnpoly , {'color': 'cyan', 'opacity': 0.5}, 'Vegetação nativa')
Map
```

```python
areatotal = int(roi.area().getInfo()) / 10000
areavn = int(vnpoly.aggregate_sum('count').getInfo()) * (100) / 10000
print('Area vegetação nativa: ', areavn)
print('Area total: ', areatotal)
print('% de Vn: ', (areavn/areatotal)*100)
```
